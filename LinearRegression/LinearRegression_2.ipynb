```python
import pandas as pd
from pandas import Series, DataFrame
import sklearn as sk
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

```

```python
df = pd.read_csv('/content/sample_data/fish_train.csv')
df
```

```python
X = df.drop(['Species', 'Weight'], axis = 1)
Y = df['Weight']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=11, stratify= df.loc[:,'Species'])
```

```python
l = LinearRegression().fit(x_train, y_train)
l.score(x_train, y_train)
```

```python
y_predicted = l.predict(x_test)
r2 = r2_score(y_test, y_predicted)
r2
```
